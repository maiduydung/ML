```python
weight = 50
goal = 0.8
input = -0.5

weight_arr = []
error_arr = []
for i in range(100):
    pred = input * weight
    error = (pred - goal)**2

    weight = weight - 1
    weight_arr.append(weight)
    error_arr.append(error)
    
    print("Error ", error, "Prediction ", pred)
```

```python
import matplotlib.pyplot as plt
plt.plot(weight_arr, error_arr)
plt.gca().invert_xaxis()
```
