### Analysis of infection based on SEIR model  

#### Import libraries  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import odeint, solve_ivp
from scipy.optimize import minimize
```

#### Parameters  

```python
dt = 0.01  # for odeint
```

- S: #Susceptible = the number of people who are not immune
- E: #Exposed = the number of people who are infected and during the incubation period
- I: #Infectious = the number of people who are onset
- R: #Removed = the number of people who have been infected and are recovered (have immunity) or dead
- beta: infectious rate (S -> E)
- lp: length of latent period (incubation period, E -> I)
- ip: length of infectious period (I -> R

```python
def seir_eq(v, t, beta, lp, ip):
    S, E, I, R = v
    dSdt = -beta * S * I
    dEdt = beta * S * I - E / lp
    dIdt = E / lp - I / ip
    dRdt = I / ip
    return dSdt, dEdt, dIdt, dRdt
```

```python
init_S = 762
init_E = 0
init_I = 1
init_R = 0
init_state = [init_S, init_E, init_I, init_R]

beta = 0.02
lp = 1.0
ip = 2.0

t_max = 14
t = np.arange(0, t_max, dt)
num_t = int(t_max/dt)
step_t = int(1/dt)
```

```python
sol = odeint(seir_eq, init_state, t, args=(beta, lp, ip))
plt.plot(t, sol)
plt.xlabel('Days')
plt.ylabel('Population')
plt.legend(['Susceptible', 'Exposed', 'Infected', 'Removed'])
plt.show()
```
