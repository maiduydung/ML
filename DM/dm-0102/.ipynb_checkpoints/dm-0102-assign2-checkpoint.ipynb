## Create simple plot with Pandas  

### Import libraries

```python
import pandas as pd
```

### Parameters

```python
csv_in1 = '../../AI/ai-0102/2045_2017-utf8.csv'
csv_in2 = '../../AI/ai-0102/1393_2017-utf8.csv'
```

### Read CSV file

```python
df1 = pd.read_csv(csv_in1, delimiter=',', skiprows=9, header=0)
df1['Date']=pd.to_datetime(df1['Date'], format="%Y-%m-%d")

df2 = pd.read_csv(csv_in1, delimiter=',', skiprows=9, header=0)
df2['Date']=pd.to_datetime(df1['Date'], format="%Y-%m-%d")
print(df1.shape)
print(df1.info())
display(df1.head())


print(df2.shape)
print(df2.info())
display(df2.head())
```

```python
df12 = pd.merge(df1, df2, on='Date', how='inner')
df12.head()
```

```python
df12.plot.scatter('AdjClose_x', 'AdjClose_y')
```
