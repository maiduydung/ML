```python
import numpy as np
import pandas as pd
import folium
```

```python
df = pd.read_csv('../crime2006_2016_en-utf8.csv')
```

```python
df = df[df['Year']==2016]
```

```python
df
```

```python
df_pos = pd.read_csv('../h3010pref_pos_utf8_en.csv', delimiter=',', skiprows=0, header=0)
print(df_pos.shape)
display(df_pos.head())
```

```python
df = pd.merge(df,df_pos, how='left')
```

```python
df.head()
```

```python
df_heat = df[['lat', 'lon', '#Arrested']]
print(df_heat.shape)
display(df_heat.head())
```

```python
from folium.plugins import HeatMap

m = folium.Map(location=[35, 140], zoom_start=4)
m.add_child(HeatMap(df_heat, radius=12))
display(m)
```
