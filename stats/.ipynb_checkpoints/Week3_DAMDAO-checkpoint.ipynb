```python
#geometric mean - avg growth rate
import numpy as np
x = np.array([1.2, 0.95, 1.1])
p = 1/len(x)
res = np.prod(x) # multiply all elements in x
print(np.power(res,p))
```

```python
#harmonic mean - avg speed
import scipy.stats as st
data = np.array([25,15])
st.hmean(data)
```
