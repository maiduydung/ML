### plot of exp function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2)
plt.plot(x, np.exp(-x), label='exp(-x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#?np.linspace
```

```python
x = np.linspace(-2, 2)
plt.plot(x, np.exp(-x**2), label='exp(-x**2)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#normal distribution

x = np.linspace(-10, 10)
std = x.std()
var = x.var()
mu = x.mean()
f_x = (1/(std*np.sqrt(2*np.pi)))*np.exp(-(x-mu)**2/(2*var))
plt.plot(x, f_x, label='f_x')
plt.grid(True)
plt.legend()
plt.show()
```
