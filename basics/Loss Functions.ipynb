```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Cross Entropy / Log Loss

##### Binary classification


```python
P = np.arange(0.0, 1.0, 0.01)

Y_1 = [1]*100
Y_0 = [0]*100

def CrossEntropy_bin(y, p):
    if y == 1:
        return -np.log(p)
    if y == 0:
        return np.log(1-p)

results_1 = []
results_0 = []

for i in range(len(Y_1)):
    results_1.append(CrossEntropy_bin(Y_1[i], P[i]))

for i in range(len(Y_0)):
    results_0.append(CrossEntropy_bin(Y_0[i], P[i]))
    
plt.plot(P, results_1)
plt.title('Log loss when Y = 1')
plt.xlabel('probability')
plt.ylabel('-log(p)')
plt.show()

plt.plot(P, results_0)
plt.title('Log loss when Y = 0')
plt.xlabel('probability')
plt.ylabel('log(1-p)')
plt.show()
```
