```python
a = np.array([[1,4],[9,1]])
print(a)
eigenvalues, eigenvectors = np.linalg.eig(a)
print(eigenvalues)
print(eigenvectors)
```

```python
A = np.array([[3,2],[3,-2]])
eigenValues, eigenVectors = np.linalg.eig(A)

idx = eigenValues.argsort()[::-1]   
eigenValues = eigenValues[idx]
eigenVectors = eigenVectors[:,idx]
print(eigenValues)
print('')
print(eigenVectors)
```
