```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
# creating initial dataframe
bridge_types = ('Arch','Beam','Truss','Cantilever','Tied Arch','Suspension','Cable')
bridge_df = pd.DataFrame(bridge_types, columns=['Bridge_Types'])
# creating instance of labelencoder
labelencoder = LabelEncoder()
# Assigning numerical values and storing in another column
bridge_df['Bridge_Types_Cat'] = labelencoder.fit_transform(bridge_df['Bridge_Types'])
bridge_df
```

```python
#remove all columns with null for simplicity
train_df2 = pd.read_csv('../datasets/water_table/water_table_train_values.csv')
train_df2 = train_df2.drop('funder', axis = 1)
train_df2 = train_df2.drop('installer', axis = 1)
train_df2 = train_df2.drop('subvillage', axis = 1)
train_df2 = train_df2.drop('public_meeting', axis = 1)
train_df2 = train_df2.drop('scheme_management', axis = 1)
train_df2 = train_df2.drop('scheme_name', axis = 1)
train_df2 = train_df2.drop('permit', axis = 1)
```

```python
#we're gonna one hot encode those columns with "object" datatype

type_dict = train_df2.dtypes.to_dict()
to_be_one_hot = []
for name, dtype in type_dict.items():
    if(dtype=='object'):
        to_be_one_hot.append(name)
        print(name)
```

```python
to_be_one_hot = np.array(to_be_one_hot)
to_be_one_hot.shape
```

```python
#one hot encode columns with "object" datatype
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
for col in train_df2.columns:
    if col in to_be_one_hot:
        print(col)
        unique_vals = train_df2[col].unique()
        train_df2[col] = labelencoder.fit_transform(train_df2[col])
```
