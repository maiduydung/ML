### Autograd

```python
x = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)
y= x+1
print(x)
print(y)
```

```python
a = torch.tensor(5.0, requires_grad=True)
b = torch.tensor(3.0, requires_grad=True)
```

```python
c = a**3
c
```

```python
#compute gradient
c.backward()
```

```python
a.grad.data #dz/da
```

```python
weights = torch.ones(4, requires_grad=True)

for epoch in range(3):
    model_output = (weights*3).sum()
    model_output.backward()
    print(weights.grad)
    weights.grad.zero_() #reset grad
```

## Backprop

```python
x = torch.tensor(1.0,requires_grad=True)
w = torch.tensor(1.0,requires_grad=True)
y = torch.tensor(3.0,requires_grad=True)

#forward pass to get the loss
y_hat = w*x
loss = (y-y_hat)**2
print('y_hat: ', y_hat)
print('loss: ', loss)

#backward pass
loss.backward()
d_loss_over_d_w = w.grad
print('d_loss/d_w: ', d_loss_over_d_w,'\n')

#new weight
lr = 0.01
w_new = w-lr*d_loss_over_d_w
loss_new = (y-w_new*x)**2
print('w_new: ', w_new)
print('loss_new: ', loss_new)
```

```python
#loop
for i in range(10):
    w_new = w-lr*d_loss_over_d_w
    w = w_new
    loss_new = (y-w_new*x)**2
    print(loss_new, w_new)
```
