```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import balanced_accuracy_score, roc_auc_score, make_scorer
from sklearn.model_selection import GridSearchCV #cross_valid
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
import datetime
```

```python
df = pd.read_csv('all_scores_updated.csv')
df['founded_on_date'] = pd.to_datetime(df['founded_on_date'], errors = 'coerce') #avoiding Out of bounds nanosecond timestamp: 1651-01-01 00:00:00
df['founded_on_year'] = pd.DatetimeIndex(df['founded_on_date']).year
df['company_age'] = (2021 - df['founded_on_year']) #so that XGBoost can group companies with similar age to gether instead of using it as a category variable 
```

```python
df
```

```python
df.dtypes
```

#### Missing data

```python
print('Length of non null cols and their percentage\n')
for col in df.columns:
    print(col, len(df[df[col].notnull()]), round(df[col].count()*100/len(df),3))
```

#### Set null to 0

```python
for col in df.columns:
    df.loc[(df[col].isna()), col] = 0
df.head()
```

#### Preparing train and test set

```python
X = df.drop(['name', 'domain','score','founded_on_date', 'founded_on_year'],axis = 1).copy()
X.head()
```

```python
Y = df['score'].copy()
Y.head()
```

```python
plt.hist(Y)
# very skewed
```

### Reducing skewness

```python
#for acc ~0.77 

df_score_1 = df[df['score']==1][:1169]
df_score_2 = df[df['score']==2][:1169]
df_score_3 = df[df['score']==3][:1169]
df_score_4 = df[df['score']==4][:1169]
df_score_5 = df[df['score']==5]
frames = [df_score_1, df_score_2, df_score_3, df_score_4, df_score_5]
```

```python
df = pd.concat(frames)
print(df.shape)
```

```python
plt.hist(df['score'])
```

#### Drop unnecessary features

```python
X = df.drop(['name', 'domain','score','founded_on_date', 'founded_on_year'],axis = 1).copy()
print(X.shape)
X.head()
```

```python
sns.heatmap(X.corr());
```

```python
Y = df['score'].copy()
Y.head()
```

### One hot encoding

```python
X_encoded = pd.get_dummies(X, columns=['categories', 
                                       'country_iso_3166_1_alpha_3', 
                                       'latest_funding_type', 
                                       'employee_count_int4range',
                                       ])
X_encoded.head()
```

### Removing [], in column names for XGBoost

```python
X_encoded.columns = X_encoded.columns.str.replace('[', '', regex=True)
X_encoded.columns = X_encoded.columns.str.replace(']', '', regex=True)
X_encoded.columns = X_encoded.columns.str.replace(',', '', regex=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, Y, stratify = Y, random_state=42)# stratify = y to maintain the ratio
```

```python
X_train.head()
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train.dtypes
```

```python
clf_xbg = xgb.XGBClassifier(objective='multi:softmax', missing=None, seed=42)
clf_xbg.fit(X_train, y_train, verbose=True, early_stopping_rounds=60, eval_metric='merror', eval_set=[(X_test, y_test)])
```

```python
%%time
y_pred=clf_xbg.predict(X_test)
```

```python
matrix = confusion_matrix(y_test, y_pred)

matrix = matrix.astype('float') / matrix.sum(axis=1)[:, np.newaxis]
sns.heatmap(matrix, annot=True, annot_kws={'size':10}, cmap=plt.cm.Greens, linewidths=0.2)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title("Confusion matrix for XGBoost")
plt.show()
```

```python
accuracy_score(y_test, y_pred, normalize=True)
```

### K-Fold

```python
data_dmatrix = xgb.DMatrix(data = X_encoded, label = Y)

#Manually adding class num to params so that xgb.cv() won't throw an error
xgb_params = clf_xbg.get_xgb_params()
xgb_params['num_class'] = 6
xgb_params['eval_metric'] = 'merror'
```

```python
xgb_params
```

```python
params = {'objective':'multi:softmax', 'colsample_bytree':0.3, 'learning_rate':0.1, 'alpha':10} #
xgb_cv_score = xgb.cv(dtrain=data_dmatrix, params=xgb_params, nfold=5, as_pandas=True)
```

```python
xgb_cv_score
```

```python
importance_dict = clf_xbg.get_booster().get_score(importance_type='gain')
```

```python
dictlist=[]
for key, value in importance_dict.items():
    temp = [key,value]
    dictlist.append(temp)
```

```python
dictlist = np.array(dictlist)
```

```python
importance_df = pd.DataFrame({'feature': dictlist[:, 0], 'importance': dictlist[:, 1]})
```

```python
importance_df.sort_values(by='importance')
```
