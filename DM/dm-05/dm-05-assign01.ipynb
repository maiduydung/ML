```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
df = pd.read_csv('name_gender_dataset.csv')
df = df.sample(frac=1) #shuffle
df = df[:10000]##using only the top 10000 otherwise we got overflow
print(df.shape)
print(df.info())
display(df.head())
```

```python
df_train = df[:7000]
df_test = df[7000:]

X_train = df_train['Name']
y_train = df_train['Gender']
print(X_train.shape, y_train.shape)

X_test = df_test['Name']
y_test = df_test['Gender']
print(X_test.shape, y_test.shape)
```

```python
vectorizer = CountVectorizer()
vectorizer.fit(X_train)
vocab = vectorizer. get_feature_names()
print('Vocabulary size:', len(vocab))
#print(vocab)
```

```python
X_train_bow = vectorizer.transform(X_train)
X_test_bow = vectorizer.transform(X_test)
print('X_train_bow:')
print(repr(X_train_bow))
print('X_test_bow:')
print(repr(X_test_bow))
```

```python
Xbow = pd.DataFrame(X_train_bow.toarray(), 
                    index=y_train, columns=vocab)
display(Xbow)
```

```python
model = MultinomialNB(alpha=1.0)
model.fit(X_train_bow, y_train)
print(model.classes_)
```

```python
train_score = model.score(X_train_bow, y_train)
print('Train accuracy:', train_score)
```

```python
# Obtain each prediction  
y_test_pred = model.predict(X_test_bow)
df_pred = pd.DataFrame({
            'pred': y_test_pred,
            'true': y_test
          }).reset_index(drop=True)
display(df_pred.head())
```

```python
proba = model.predict_proba(X_test_bow)
results = pd.DataFrame(proba, columns=model.classes_)
print('Prediction:')
display(results)
print(model.predict(X_test_bow))
```
