## Example of displaying heatmap using folium (English)  

### Import libraries  

```python
import pandas as pd
import folium
```

### Parameters

```python
csv_in = 'crime2006_2016_en-utf8.csv'
geo = 'japan.json'
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

### Read CSV file

```python
df = pd.read_csv(csv_in, delimiter=',', skiprows=0, header=0)
print(df.shape)
print(df.info())
display(df.head())
#display(df)
```
