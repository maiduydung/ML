```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Annual growth

```python
x = 1
for i in range(3):
    x_t_minus_1 = x
    x = x+x*1/1
    print('x = {0}\nx_t_minus_1 = {1}\nx/x_t_minus_1 = {2}\n'.format(x, x_t_minus_1, x/x_t_minus_1))
```

### Semi annual growth

```python
x = 1
for i in range(3):
    x_t_minus_1 = x
    x = x+x*1/2
    print('x = {0}\nx_t_minus_1 = {1}\nx/x_t_minus_1 = {2}\n'.format(x, x_t_minus_1, x/x_t_minus_1))
```

### Quarterly growth

```python
x = 1
for i in range(5):
    x_t_minus_1 = x
    x = x+x*1/4
    print('x = {0}\nx_t_minus_1 = {1}\nx/x_t_minus_1 = {2}\n'.format(x, x_t_minus_1, x/x_t_minus_1))
```

#### only check the end result: x_n

```python
n = np.arange(1,100)
```

```python
res = []
for ele in n:
    #print((1+1/ele)**ele)
    res.append((1+1/ele)**ele)
#print(res)
```

```python
df_res = pd.DataFrame(list(zip(n, res)),columns =['n', 'result'])
df_res
```

```python
plt.plot(df_res['n'], df_res['result'])
plt.xlabel('n')
plt.ylabel('result')
```
