```python
import pandas as pd
import numpy as np
```

```python
h_w = pd.read_csv('datasets/height_weight/height_weight.csv')


# Since height is in pound and weight is in inch.
# We're doing a bit conversion here

h_w_np = np.array(h_w)
# Inch to M
h_w_np[:,1] = np.dot(h_w_np[:,1], 0.0254)
h_w_np[:,2] = np.dot(h_w_np[:,2], 0.45359237)

#covert back to pd dataframe
pd.DataFrame()
```
