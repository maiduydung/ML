```python
# Directly finding out the theta that minimizes the lost fuction
# Works well with small number of features
import numpy as np
import matplotlib.pyplot as plt

#Input
X = 2 * np.random.rand(100,1)

# Dummy equation
y = 3*X + 4 + np.random.randn(100, 1)

plt.scatter(X,y)
```

```python
#X_bias

# Always use .dot() from now on
# The extra column of ones is the "input" to the intercept term
X_b = np.c_[np.ones((100,1)), X]
X_b[:4]
```

```python
first_half = np.linalg.inv(X_b.T.dot(X_b))
first_half
```

```python

second_half = X_b.T.dot(y)
second_half
```

```python
theta_best = first_half.dot(second_half)
theta_best

# θ_0 = 3.90042826
# θ_1 = 2.89115787
# y_hat = 3.90042826 + 2.89115787*X

# y = 4 + 3*X + Gaussian noise
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2,1)), X_new] #adding x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-") #red dash
plt.plot(X, y, "b.") #blue dot
plt.axis([0, 2, 0, 15]) #xmin, xmax, ymin, ymax
plt.show()
```
