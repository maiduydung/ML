```python
a = np.array([1,3,8,2,4])
b = np.roll(a,1)
print("a:",a)
print("b:",b)
print("abs(a-b)",abs(a-b))
print("a, b one on one difference", abs(a-b).sum())

distance, path = fastdtw(a, b)
print("DTW difference", distance)
print(path)
```
